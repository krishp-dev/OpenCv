removing R G B

```python
img = cv2.imread('download.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_no_red = img_rgb.copy()
img_no_green = img_rgb.copy()
img_no_blue = img_rgb.copy()

img_no_red[:,:,0] = 0
img_no_green[:,:,1] = 0
img_no_blue[:,:,2] = 0

images_array = np.array([img_no_red, img_no_green, img_no_blue])

titles = ['No Red Channel', 'No Green Channel', 'No Blue Channel']
plt.figure(figsize=(15, 5))

for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images_array[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

print("Shape of the numpy array:", images_array.shape)
```

fliping a image

```python
img = cv2.imread('download.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

flip_0 = cv2.flip(img_rgb, 0)  
flip_1 = cv2.flip(img_rgb, 1)  
flip_minus1 = cv2.flip(img_rgb, -1)  
flipped_images = np.array([img_rgb, flip_0, flip_1, flip_minus1])

titles = ['Original', 'Vertical Flip (0)', 'Horizontal Flip (1)', 'Both Directions (-1)']
plt.figure(figsize=(15, 10))

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(flipped_images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

print("Shape of flipped images array:", flipped_images.shape)
```

Cropping Image

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.cvtColor(cv2.imread('download.jpeg'), cv2.COLOR_BGR2RGB)
cropped_img = img[50:300, 100:400]

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original')

plt.subplot(1, 2, 2)
plt.imshow(cropped_img)
plt.title('Cropped')

plt.show()
```

Saving Image

```python

img = cv2.imread('download.jpeg')
cropped_img = img[50:300, 100:400]
cv2.imwrite('cropped_image.jpg', cropped_img)
```
